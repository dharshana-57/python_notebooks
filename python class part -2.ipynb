```python
#writing contents in file
from random import randint

def main():
    #open file for writing data
    outfile = open("inputfile.txt","w")
    for i in range(10):
        outfile.write(str(randint(0, 9)) + " ")
    outfile.close()
    #open file for reading data
    infile = open("inputfile.txt", "r")
    s = infile.read()
    numbers = [eval(x) for x in s.split()]
    for n in numbers:
        print(n,end = " ")
    infile.close()
    
main()
```

```python
# Exception Handling 
# to avoid abrupt termination 
# continues normal execution of program
# syntax:-
# try:
#    <error code>
# except <Exception type>:
#   <handler>
# single try block can have many except classes
# try:
#    <error code>
# except <Exception type1>:
#   <handler>
# except <Exception type2>:
#   <handler>
# except <Exception type3>:
#   <handler>
# ....
# else:
#    statement 
# finally:
#    statement 
# else statement  if there is no exception 
# finally -> any one of except block executed  or else part execute after execute
```

```python
# Exception Handling to check file exists or not
def main():
    while True:
        try:
            filename = input("Enter filename:").strip() #.strip() - extra spaces will be removed
            infile = open(filename,"r")
            break
        except IOError:
            print("file entered doesn't exists")
            
    print("file exists!")
    
    
main()
```

```python
# one try block with many except clauses
def main():
    while True:
        try:
            num1, num2 = eval(input("Enter 2 number:"))
            r = num1 / num2;
            print(r)
            
        except ZeroDivisionError:
            print("invalid! number not divisible by zero")
        except SyntaxError:
            print("check Syntax")   
        else:
            print("No Exception")
        finally:
            print("Final clause!")
    
    
main()
```

```python
# python set operations -ass 1)
s1 = {1,3,"hello","a","4"}
s2 = {2,4,"hello", "4",3,90}
print("s1 =:",s1)
print("s2 =",s2)
print("Union of s1 and s2 =", s1 | s2)
print("Intersection of s1 and s2 =", s1 & s2)
print("Difference of s1 and s2 =", s1 - s2)
print("Symmetric difference of s1 and s2 =",s1 ^ s2)
```

```python
# functions returns objects or instances 
# objects can be passed as argument to a function
class test1:
    #attributes,member function
    pass

ob1 = test1() # object creation

def area(ob3):
    ob2 = test1()
    ob2.a = ob3.x * ob3.y
    return ob2.a  # returning objects

ob1.x = 20
ob1.y = 30
findarea = area(ob1) # passing objects as arguments
print("Area of Rectangle is %g" % findarea)
```

```python
def sum_n(*args):
    global sum
    sum = 0
    for n in args:
        if n<0:
            raise ValueError("You have Entered Negative Number!")
            break
        try:
            if n>=0:
                sum+=n
                print("sum after adding",n,":",sum)
        except ValueError as e:
            print(e)
            
    
sum_n(4,5,6,7,-8)
```
